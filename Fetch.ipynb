```python
from pathlib import Path

def get_package(dataset_id: str, folder="tmp", max_size=10000):
    get_package_url=f"{essdive_api_url}/packages/{dataset_id}"
    dataset_response = requests.get(get_package_url,
                          headers={"Authorization": f"Bearer {token}"})
    info = dataset_response.json()
    print(list(info.keys()))
    for d in info["dataset"]["distribution"]:
        url = d["contentUrl"]
        name = d["name"]
        size = d["contentSize"]
        print(name, size, url)
        if size > max_size:
            continue
        mode = 'w'
        path = Path(folder) / dataset_id / name
        path.parent.mkdir(parents=True, exist_ok=True)
        dist_response = requests.get(url)
        if d["encodingFormat"] == "text/csv":
            with open(str(path), "w", encoding='utf-8') as file:
                file.write(dist_response.text)
        else:
            with open(str(path), "wb") as file:
                file.write(dist_response.content)
    return info

get_package('ess-dive-a9de0ff71198b5e-20231024T150445803')
```
